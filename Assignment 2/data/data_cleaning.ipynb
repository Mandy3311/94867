```python
import numpy as np
import pandas as pd
```

```python
county = pd.read_csv('covid_county_population_usafacts.csv')
confirmed = pd.read_csv('covid_confirmed_usafacts.csv')
```

```python
df = pd.merge(county, confirmed, on='State')
df = df.iloc[:, :352]
df = df.drop(['countyFIPS_x', 'countyFIPS_y', 'County Name_y', 'StateFIPS'], axis=1)
df.sample(5)
```

```python
state_group = df.groupby('State').sum()
```

```python
state_group.sample(5)
```

```python
for col in range(1, 346):
    state_group.iloc[:, col] = state_group.iloc[:, col] / state_group['population']
state_group = state_group.drop(['population'], axis=1)
```

```python
df = state_group.T
df.sample(5)
```

```python
df = df.reset_index()
df = df.rename(columns={'index': 'date'})
```

```python
df.sample(5)
```

```python
file_path = 'cleaned.csv'
df.to_csv(file_path, index=False)
```
